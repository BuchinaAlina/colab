# 2.Изучите стоимости недвижимости

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('kc_house_data.csv', sep=',')
df.head()
```

```python
df['price'].min(), df['price'].max()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['price'] / 1e6, bins=15)
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Цена (в миллионах)')
plt.ylabel('Количество домов')
plt.xticks(np.arange(0.05, 5.0, 0.5), range(0, 50, 5), rotation=45)
plt.yticks(range(0, 16000 , 1000))
plt.grid()
;
```

Из гистограммы видно, что большинство домов находятся в пределах стоимости от 1 до 10 миллионов. Причем около 14900 домов находятся в диапазоне цен от 1 до 5 миллионов, а 5500 домов нахдятся от 5 до 10 миллионов.


UPD: Гистограмма немного не корректоно отобразилась.

# 3. Изучите распределение квадратуры жилой площади

```python
plt.figure(figsize=(10, 6))
plt.hist(df['sqft_living'], bins=60)
plt.title('Распределение квадратуры жилой площади')
plt.xlabel('кв.м')
plt.ylabel('Количество')
plt.xticks(range(100, 9500, 1000), rotation=45)
plt.yticks(range(0, 2800, 300))
plt.grid()
;
```

# 4. Изучите распределение года постройки


```python
plt.figure(figsize=(16, 6))
plt.hist(df['yr_built'], bins=40)
plt.title('Распределение года постройки')
plt.xlabel('год постройки')
plt.ylabel('Количество')
plt.xticks(range(1900, 2020, 5), rotation=45)
plt.yticks(range(0, 1500, 50))
plt.grid()
;
```

# Задание 3
## Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='waterfront', y='price', data=df)
plt.title('Зависимость цены от наличия вида на набережную')
plt.xlabel('Вид на набережную')
plt.ylabel('Цена')
plt.ticklabel_format(style='plain', axis='y')
plt.grid()
;

```

В данном боксплоте отображена зависимость цены  от наличия вида на набережную, где 0-нет вида на набережную, 1-есть вид на набережную. Таким образом цена домов выше у домов с видом на набережную. Средняя цена таких домов составляет примерно 1500000. Но также отмечается наличие домов, цена которых совпадает с ценой домов без вида на набережную.

```python
plt.figure(figsize=(16, 6))
sns.boxplot(x='view', y='price', data=df)
plt.title('Зависимость цены от вида')
plt.xlabel('Вид')
plt.ylabel('Цена')
plt.ticklabel_format(style='plain', axis='y')
plt.grid()
;
```

В данном блокплоте значение от 0 до 4 насколько хороший вид. Отмечается соответствие повышения цены в зависимости от вида. Примечательно, что дома с оценкой вида 1 и 2 находятся в одном ценовом диапазоне.

```python
plt.figure(figsize=(16, 6))
sns.boxplot(x='condition', y='price', data=df)
plt.title('Зависимость цены от состояния')
plt.xlabel('Состояние')
plt.ylabel('Цена')
plt.ticklabel_format(style='plain', axis='y')
plt.grid()
;
```

 В данном блокплоте начение от 1 до 5 насколько хорошее состояние. Дома с оценкой состояния 1 и 2 находятся в одном ценовом диапазоне. А у домов с оценкой 3,4 и 5 цена незначительно выше и они тоже находятся примерно в одном ценовом диапазоне.

```python
plt.figure(figsize=(16, 6))
sns.boxplot(x='grade', y='price', data=df)
plt.title('Зависимость цены от уровня конструкции и дизайна')
plt.xlabel('Уровень конструкции и дизайн')
plt.ylabel('Цена')
plt.ticklabel_format(style='plain', axis='y')
plt.grid()
;
```

Приведенный блокплот отображает зависимость цены от уровня конструкции и дизайна домов. Значение от 1 до 13, где 1-3 плохая конструкция здания и дизайн, 4-7 - средний уровень конструкции и дизайна, 11-13 - высокое качество конструкции и дизайна. В данном случае хорошо заметно, чем выше оценка, тем выше цена.

```python
plt.figure(figsize=(16, 6))
sns.boxplot(x='bathrooms', y='price', data=df)
plt.title('Зависимость цены от количества ванных комнат')
plt.xlabel('Количество ванных комнат')
plt.ylabel('Цена')
plt.ticklabel_format(style='plain', axis='y')
plt.grid()
;
```

В этом  блокплоте также цена дома увеличивается соответственно количеству ванных комнат.
